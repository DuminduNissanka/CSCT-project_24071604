```python
import pandas as pd
import requests
```

```python
# Bristol city centre coordinates
LAT = 51.4545
LON = -2.5879
```

```python
start_date = "2024-01-01"
end_date   = "2024-12-31"
```

```python
#get weather from API
url = (
    "https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={LAT}&longitude={LON}"
    f"&start_date={start_date}&end_date={end_date}"
    "&daily=temperature_2m_max,temperature_2m_min,temperature_2m_mean,"
    "precipitation_sum,rain_sum,snowfall_sum,windspeed_10m_max,windspeed_10m_mean"
    "&timezone=Europe/London"
)

response = requests.get(url)
data = response.json()

weather = pd.DataFrame(data["daily"])
weather["date"] = pd.to_datetime(weather["time"])
weather = weather.drop(columns=["time"])

weather.to_csv("C:/Users/DuminduS/Desktop/UWE/CSCT Masters Project/datasets/bristol_weather_2024.csv", index=False)

weather.head()
```
