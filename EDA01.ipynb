```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    "C:/Users/DuminduS/Desktop/UWE/CSCT Masters Project/datasets/Final data/Final02.csv"
)

df["date"] = pd.to_datetime(df["date"], format="mixed", dayfirst=True)
```

```python
plt.figure()
sns.histplot(df["sessions"], bins=30, kde=True)
plt.xlabel("Daily Charging Sessions")
plt.ylabel("Frequency")
plt.title("Distribution of Daily EV Charging Sessions")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df["sessions"])
plt.xlabel("Daily Charging Sessions")
plt.title("Boxplot of Daily EV Charging Sessions")
plt.show()
```

```python
plt.figure()
sns.histplot(df["energy_kwh"], bins=30, kde=True)
plt.xlabel("Daily Energy Delivered (kWh)")
plt.ylabel("Frequency")
plt.title("Distribution of Daily Energy Consumption")
plt.show()
```

```python
plt.figure()
sns.histplot(df["mean_charge_duration_min"], bins=30, kde=True)
plt.xlabel("Mean Charging Duration (minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of Average Charging Duration")
plt.show()
```

```python
plt.figure()
sns.countplot(x=df["n_connectors"])
plt.xlabel("Number of Connectors")
plt.ylabel("Count")
plt.title("Distribution of Charging Station Capacity")
plt.show()
```

```python
plt.figure()
sns.histplot(df["temperature_2m_mean"], bins=30, kde=True)
plt.xlabel("Mean Daily Temperature (C)")
plt.ylabel("Frequency")
plt.title("Distribution of Mean Daily Temperature")
plt.show()
```

```python
plt.figure()
sns.histplot(df["precipitation_sum"], bins=30, kde=True)
plt.xlabel("Daily Precipitation (mm)")
plt.ylabel("Frequency")
plt.title("Distribution of Daily Precipitation")
plt.show()
```

```python
plt.figure()
sns.histplot(df["windspeed_10m_mean"], bins=30, kde=True)
plt.xlabel("Mean Wind Speed (m/s)")
plt.ylabel("Frequency")
plt.title("Distribution of Mean Wind Speed")
plt.show()
```

```python
plt.figure()
sns.histplot(df["avg_daily_traffic_nearby"], bins=30, kde=True)
plt.xlabel("Average Daily Traffic Near Station")
plt.ylabel("Frequency")
plt.title("Distribution of Traffic Intensity Near Charging Stations")
plt.show()
```

```python
plt.figure(figsize=(8,3))
df["date"].value_counts().sort_index().plot()
plt.xlabel("Date")
plt.ylabel("Number of Station Records")
plt.title("Temporal Coverage of the Dataset")
plt.show()
```
